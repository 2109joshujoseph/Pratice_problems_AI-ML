```python
# Perform meaningful visualizations using the dataset. Bring at least 3
#visualizations.
#Bar Plots of total medals by  country
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df['Total'] = df['Gold'] + df['Silver'] + df['Bronze']
plt.figure(figsize=(10, 6))
sns.barplot(x='Unnamed: 0', y='Total', data=df)
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.title('Total Medals by Country')
```

```python
#pie-chart of medal distribution
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
labels = ['Gold', 'Silver', 'Bronze']
sizes = [df['Gold'].sum(), df['Silver'].sum(), df['Bronze'].sum()]
colors = ['gold', 'silver', 'brown']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Medal Distribution')
plt.axis('equal')
```

```python
#Scatter Plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unnamed: 0', y='Total', data=df)
```
